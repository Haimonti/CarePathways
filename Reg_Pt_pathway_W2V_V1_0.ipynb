```python
X = df.drop(['id', 'Hospital LOS', 'Mean HLOS', 'Std Dev'], axis = 1)
```

# Model Training

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 42, shuffle = True) # did not use stratify as this is not a classification problem
```

```python
print(X_train.shape, X_test.shape, y_train.shape)
```

```python
regressor = LinearRegression()
```

```python
X_pred = regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
print(y_pred)
```

```python
r2_test_error = metrics.r2_score(y_test, y_pred)
print('R squared error for testing data:', r2_test_error)
```
