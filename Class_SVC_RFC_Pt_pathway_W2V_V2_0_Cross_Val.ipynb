````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
import os

from google.colab import drive
from pathlib import Path
drive.mount("/content/drive")
````

````python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/HMCPW/data/v2/pt_pathways-w2v-top200-v2.csv')
````

````python
df.shape
````

````python
df.head()
````

````python
df = df.drop(['id', 'Hospital LOS', 'Std Dev', 'Cancer'], axis = 1)
````

````python
df = df.drop(['reason_for_admission_Atypical pneumonia [J18.9]', 'reason_for_admission_Fever of unknown origin [R50.9]','reason_for_admission_Hypoxic [R09.0]','reason_for_admission_Myalgia [M79.19]','reason_for_admission_Tachypnea [R06.8]', 'reason_for_admission_Viral pneumonia [J12.9]'], axis = 1)
````

````python
df.shape
````

````python
df.head()
````

````python
X = df.drop(['Mean HLOS'], axis = 1)
y = df[['Mean HLOS']]
````

````python
X
````

````python
y
````

# Logistic


````python
model = LogisticRegression(max_iter=1000)
````

````python
# CV just for trainig
cv_score_lr = cross_val_score(model, X, y, cv = 10)
````

````python
print(cv_score_lr)
````

````python
mean_cv_score_lr = sum(cv_score_lr)/len(cv_score_lr)
mean_cv_score_lr = mean_cv_score_lr*100
mean_cv_score_lr = round(mean_cv_score_lr, 2)
print(mean_cv_score_lr)
````

# SVM *Classifier*

````python
from sklearn.svm import SVC
````

````python
#Accuracy in test data
cv_score_svc = cross_val_score(SVC(kernel='linear'), X, y, cv = 10)
````

````python
print(cv_score_svc)
````

````python
mean_cv_score_svc = sum(cv_score_svc)/len(cv_score_svc)
mean_cv_score_svc = mean_cv_score_svc*100
mean_cv_score_svc = round(mean_cv_score_svc, 2)
print(mean_cv_score_svc)
````

# KNN

````python
from sklearn.neighbors import KNeighborsClassifier
````

````python
cv_score_knn = cross_val_score(KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2), X, y, cv = 10)
````

````python
print(cv_score_knn)
````

````python
mean_cv_score_knn = sum(cv_score_knn)/len(cv_score_knn)
mean_cv_score_knn = mean_cv_score_knn*100
mean_cv_score_knn = round(mean_cv_score_knn, 2)
print(mean_cv_score_knn)
````
